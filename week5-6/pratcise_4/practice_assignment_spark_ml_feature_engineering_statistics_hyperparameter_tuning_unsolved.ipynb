
# Exercise 4: Create Customer Dataset

## Background


A telecom company stores customer information.

The dataset will be reused throughout the notebook.


## Dataset / Input


Columns

customer_id
age
monthly_bill
city
churn


## Task


Create a Spark DataFrame.


## Hint


Use spark.createDataFrame().



```python

# =====================================================
# TODO
# =====================================================


# Create DataFrame
fake = Faker('en_IN')
customer_data = []
for i in range(1000):
    customer_id = 1000 + i
    age = random.randint(18, 70)
    monthly_bill = round(random.uniform(30.00, 150.00), 2)
    city = random.choice(["Chennai", "Mumbai", "Bengaluru", "Delhi", "Hyderabad"])
    churn = random.choice(["Yes", "No"])
    customer_data.append((customer_id, age, monthly_bill, city, churn))

columns = ['Customer_ID', 'Age', 'Monthly_Bill', 'City', 'Churn' ]
customer_df = spark.createDataFrame(customer_data, columns)
customer_df.show(5)
```
