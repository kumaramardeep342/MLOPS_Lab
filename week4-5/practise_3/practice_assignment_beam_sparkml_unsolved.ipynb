# Exercise 3: Create Your First Pipeline

## Background


An online store wants to process product names.

The first step is to create a pipeline and
place the data into a PCollection.


## Dataset / Input


Products

Laptop
Phone
Tablet
Monitor


## Task


Create a pipeline that prints all products.


## Hint


Use:

beam.Pipeline()
beam.Create()
beam.Map(print)



```python
# =====================================================
# TODO
# =====================================================
# Create the pipeline
products = [
    "Laptop",
    "Phone",
    "Tablet",
    "Monitor"
]
with beam.Pipeline(options=PipelineOptions()) as pipeline:
    # Create a PCollection from the list of products
    product_pcollection = pipeline | "Create Product PCollection" >> beam.Create(products)
    
    # Print the products in the PCollection
    product_pcollection | "Print Products" >> beam.Map(print)
```


# Exercise 5: Map Transformation

## Background


Product names are stored in lowercase.

Management wants all names converted to uppercase.


## Dataset / Input


Products

laptop
phone
tablet
monitor


## Task


Convert all products to uppercase.


## Hint


Use beam.Map().
Use str.upper().



```python

# =====================================================
# TODO
# =====================================================


# Convert to uppercase
products = [
    "laptop",
    "phone",
    "tablet",
    "monitor"
]
with beam.Pipeline(options=PipelineOptions()) as pipeline:
    product_pcollection = pipeline | "Create Product PCollection" >> beam.Create(products)
    
    # Convert products to uppercase
    uppercase_products = product_pcollection | "Convert to Uppercase" >> beam.Map(lambda x: x.upper())
    
    # Print the uppercase products
    uppercase_products | "Print Uppercase Products" >> beam.Map(print)

```


# Exercise 6: Filter Transformation

## Background


The inventory team only wants products having
more than six characters.


## Dataset / Input


Products

Laptop
Phone
Tablet
Keyboard
Monitor


## Task


Filter products whose length is greater than 6.


## Hint


Use beam.Filter().
Use len().



```python

# =====================================================
# TODO
# =====================================================


# Filter products
products = [
    "Laptop",
    "Phone",
    "Tablet",
    "Monitor",
    "Keyboard"
]
with beam.Pipeline(options=PipelineOptions()) as pipeline:
    product_pcollection = pipeline | "Create Product PCollection" >> beam.Create(products)
    
    # Filter products that have length greater than 6
    filtered_products = product_pcollection | "Filter Products" >> beam.Filter(lambda x: len(x) > 6)
    
    # Print the filtered products
    filtered_products | "Print Filtered Products" >> beam.Map(print)
    
# to visaulize the pipeline structure
# proto = pipeline.to_runner_api()
# print(proto)
```


# Exercise 7: FlatMap Transformation

## Background


Website clickstream logs are stored as sentences.

Each sentence contains multiple words.

FlatMap converts one element into many outputs.


## Dataset / Input


Logs

"user login successful"

"product added cart"

"checkout completed"


## Task


Split every sentence into individual words.


## Hint


Use:

beam.FlatMap()
str.split()



```python

# =====================================================
# TODO
# =====================================================


# Split sentences
logs = [
    "user login successful",
    "product added cart",
    "checkout completed"
]
with beam.Pipeline(options=PipelineOptions()) as pipeline:
    log_pcollection = pipeline | "Create Log PCollection" >> beam.Create(logs)
    
    # Split sentences into words
    words = log_pcollection | "Split Sentences" >> beam.FlatMap(lambda x: x.split())
    
    # Print the words
    words | "Print Words" >> beam.Map(print)
```

Return a tuple.



```python

# =====================================================
# TODO
# =====================================================


# Create tuples

products = [
    "Laptop",
    "Phone",
    "Tablet",
    "Monitor"
]
with beam.Pipeline(options=PipelineOptions()) as pipeline:
    product_pcollection = pipeline | "Create Product PCollection" >> beam.Create(products)
    
    # Create tuples of (product, length of product name)
    product_tuples = product_pcollection | "Create Tuples" >> beam.Map(lambda x: (x, len(x)))
    
    # Print the tuples
    product_tuples | "Print Tuples" >> beam.Map(print)
```
