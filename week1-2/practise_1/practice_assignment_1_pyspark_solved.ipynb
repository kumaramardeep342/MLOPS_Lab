
# 2. Create Spark Session

## Task
Create a SparkSession object named:
- PracticeAssignment1

The SparkSession will initialize the PySpark environment.


```python

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PracticeAssignment1") \
    .getOrCreate()

print("Spark Session Created")
```


# 3. Load Employee Dataset

## Task
Load the employee dataset using:
- spark.read.csv()

Requirements:
- Use header=True
- Use inferSchema=True
- Display first 5 rows


```python

employee_df = spark.read.csv(
    "/content/dummy_employee_dataset.csv",
    header=True,
    inferSchema=True
)

employee_df.show(5)
```


# 4. Create Department Dataset

## Task
Create a department DataFrame manually.

Columns required:
- department
- office_block
- manager_name

Display the created DataFrame.


```python

department_data = [
    ("Engineering", "A Block", "Ravi Kumar"),
    ("HR", "B Block", "Sneha Sharma"),
    ("Finance", "C Block", "Arjun Mehta"),
    ("Marketing", "D Block", "Priya Nair"),
    ("Sales", "E Block", "Vikram Rao")
]

department_df = spark.createDataFrame(
    department_data,
    ["department", "office_block", "manager_name"]
)

department_df.show()
```


# 5. Display Schema

```python

employee_df.printSchema()
```


# 6. Count Total Rows

## Task
Count total number of rows present in employee_df.


```python

employee_df.count()
```


# 7. Select Specific Columns

## Task
Display only:
- employee_id
- salary
- department


```python

employee_df.select(
    "employee_id",
    "salary",
    "department"
).show()
```


# 8. Filter Employees With Salary Greater Than 80000

## Task
Filter employees satisfying:
- salary > 80000

Display all matching rows.


```python

employee_df.filter(
    employee_df.salary > 80000
).show()
```


# 9. Filter Employees From Chennai With Experience Greater Than 5 Years

## Task
Display employees satisfying BOTH conditions:
- city = Chennai
- experience_years > 5


```python

employee_df.filter(
    (employee_df.city == "Chennai") &
    (employee_df.experience_years > 5)
).show()
```


# 10. Count Employees Per Department

## Task
Using groupBy():
- Count number of employees in each department
- Rename output column as employee_count


```python

from pyspark.sql.functions import count

employee_df.groupBy("department") \
    .agg(count("*").alias("employee_count")) \
    .show()
```


# 11. Average Salary Per Department

## Task
Find:
- Average salary for each department

Rename the output column as:
- average_salary


```python

from pyspark.sql.functions import avg

employee_df.groupBy("department") \
    .agg(avg("salary").alias("average_salary")) \
    .show()
```


# 12. Sort Employees By Salary

## Task
Sort employees:
- By salary
- In descending order


```python

employee_df.orderBy(
    employee_df.salary.desc()
).show()
```


# 13. Create Bonus Column

## Task
Create a new column named:
- bonus

Condition:
- bonus = 10% of salary


```python

from pyspark.sql.functions import col

employee_df = employee_df.withColumn(
    "bonus",
    col("salary") * 0.10
)

employee_df.show()
```


# 14. Handle Missing Values

## Task
Replace missing values in:
- city column

Replacement value:
- Unknown


```python

employee_df.na.fill({
    "city": "Unknown"
}).show()
```


# 15. Spark SQL Queries

## Task
Perform the following:
1. Create a temporary SQL view named employees
2. Write a Spark SQL query to find:
   - Average salary per department


```python

employee_df.createOrReplaceTempView("employees")

spark.sql("""
SELECT department, AVG(salary) AS avg_salary
FROM employees
GROUP BY department
""").show()
```


# 16. Find Highest Salary Employee


# 20. Inner Join Between Employee and Department Datasets

## Task
Perform:
- Inner join between employee_df and department_df

Join column:
- department


```python

joined_df = employee_df.join(
    department_df,
    on="department",
    how="inner"
)

joined_df.show()
```


# 21. Select Columns After Join

## Task
From joined_df display:
- employee_id
- department
- office_block
- manager_name
- salary


```python

joined_df.select(
    "employee_id",
    "department",
    "office_block",
    "manager_name",
    "salary"
).show()
```


# 22. Average Salary Per Office Block

## Task
Using joined_df:
- Find average salary for each office block


```python

from pyspark.sql.functions import avg

joined_df.groupBy("office_block") \
    .agg(avg("salary").alias("average_salary")) \
    .show()
```


# 23. Count Employees Per Office Block

## Task
Using joined_df:
- Count employees office-block-wise


```python

joined_df.groupBy("office_block") \
    .count() \
    .show()
```


# 24. Left Join Example

## Task
Perform:
- Left join between employee_df and department_df


```python

left_join_df = employee_df.join(
    department_df,
    on="department",
    how="left"
)

left_join_df.show()
```


# 25. Right Join Example

## Task
Perform:
- Right join between employee_df and department_df


```python

right_join_df = employee_df.join(
    department_df,
    on="department",
    how="right"
)

right_join_df.show()
```
