
# 19. Build a Simple DAG

## Task

Create a DAG containing:

- start
- task_1
- task_2
- end

using dependency chaining.


```python

# Write your code here
from airflow import DAG
from airflow.operators.empty import EmptyOperator
from datetime import datetime

with DAG(
    dag_id="simple_dag",
    start_date=datetime(2025, 1, 1),
    schedule="@daily",
    catchup=False
) as dag:

    start = EmptyOperator(task_id="start")

    task_1 = EmptyOperator(task_id="task_1")

    task_2 = EmptyOperator(task_id="task_2")

    end = EmptyOperator(task_id="end")

    start >> task_1 >> task_2 >> end
```


# 29. Conditional Task Execution

## Task

Create the following workflow:

A -> B

If B succeeds:
    Execute D

If B fails:
    Execute C

Use Airflow Trigger Rules.


```python

from airflow.utils.trigger_rule import TriggerRule

with DAG(
    dag_id="conditional_dag",
    start_date=datetime(2025,1,1),
    schedule=None,
    catchup=False
) as dag:

    task_a = EmptyOperator(task_id="task_a")

    task_b = EmptyOperator(task_id="task_b")

    task_c = EmptyOperator(
        task_id="task_c",
        trigger_rule=TriggerRule.ONE_FAILED
    )

    task_d = EmptyOperator(
        task_id="task_d",
        trigger_rule=TriggerRule.ALL_SUCCESS
    )

    task_a >> task_b
    task_b >> task_c
    task_b >> task_d

dag
```


# 36. Complex DAG Scheduling

## Task

Design the following DAG:

        A
      /   \
     B     C
      \   /
        D
        |
        E

Requirements:

- D executes only if both B and C succeed.
- E executes regardless of D success or failure.

Implement using Airflow operators and trigger rules.


```python

from airflow.utils.trigger_rule import TriggerRule

with DAG(
    dag_id="complex_dependency_dag",
    start_date=datetime(2025,1,1),
    schedule=None,
    catchup=False
) as dag:

    A = EmptyOperator(task_id="A")
    B = EmptyOperator(task_id="B")
    C = EmptyOperator(task_id="C")

    D = EmptyOperator(
        task_id="D",
        trigger_rule=TriggerRule.ALL_SUCCESS
    )

    E = EmptyOperator(
        task_id="E",
        trigger_rule=TriggerRule.ALL_DONE
    )

    A >> [B, C]
    [B, C] >> D
    D >> E

dag
```
